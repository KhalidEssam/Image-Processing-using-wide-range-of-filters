```python
def Alpha_trimmed_mean_filter(data, filter_size):
    temp = []
    indexer = filter_size // 2
    data_final = []
    data_final = np.zeros_like(data)
    for i in range(len(data)):

        for j in range(len(data[0])):

            for z in range(filter_size):
                if i + z - indexer < 0 or i + z - indexer > len(data) - 1:
                    for c in range(filter_size):
                        temp.append(0)
                else:
                    if j + z - indexer < 0 or j + indexer > len(data[0]) - 1:
                        temp.append(0)
                    else:
                        for k in range(filter_size):
                            temp.append(data[i + z - indexer][j + k - indexer])

            temp.sort()
            # print(temp)
            
            h = sum(temp[ len(temp) // 2 -filter_size : len(temp) // 2 + filter_size  ])/(2*filter_size)
            data_final[i][j] = h
            # print(data_final[i][j])
            temp = []
    return data_final
```

```python
arr=cv2.imread('R (1).png',0)
removed_noise = Alpha_trimmed_mean_filter(arr, 5)
# cv2.imshow('kj',removed_noise)
# cv2.waitKey(0)
import numpy as np
from matplotlib import pyplot as plt
fig, ax = plt.subplots()
im = ax.imshow(arr, extent=[0, 300, 0, 300],cmap = 'gray')
x = np.array(range(300))
plt.show()
```
